```python
import pandas as pd

# Load your CSV files into dataframes
df_price = pd.read_csv('Static features separate CSVs/price_action_features.csv')
df_geometry = pd.read_csv('Static features separate CSVs/geometry_features.csv')
df_tech = pd.read_csv('Static features separate CSVs/tech_features.csv')
df_patterns = pd.read_csv('Static features separate CSVs/patterns_features.csv')

# Check their columns â€” identify the common key column(s) to merge on
print(df_price.columns)
print(df_geometry.columns)
print(df_tech.columns)
print(df_patterns.columns)

# Assume all share a common key column named 'time' or 'timestamp' or 'index'
# Adjust the key name accordingly
key = 'date'  # replace with your actual common column

# Make sure the key columns are consistent type
for df in [df_price, df_geometry, df_tech, df_patterns]:
    df[key] = pd.to_datetime(df[key])  # or use appropriate conversion if not datetime

# Merge stepwise on the key column, using 'inner' join to keep only matching rows
merged_df = df_price.merge(df_geometry, on=key, how='inner') \
                    .merge(df_tech, on=key, how='inner') \
                    .merge(df_patterns, on=key, how='inner')


# Check the merged dataframe shape and info to confirm no unexpected NaNs
print(merged_df.shape)
print(merged_df.info())
print(merged_df.head())

# Save the merged dataframe to CSV
merged_df.to_csv('static_features_merged.csv', index=False)
```
